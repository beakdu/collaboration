```python
import pandas as pd
```

```python
cols=['Machine','Country','Price','Brand']
data_lst=[['TV','Korea',1000,'A'],['TV','Japan',1300,'B'],['TV','China',300,'C'],['PC','Korea',2000,'A'],['PC','Japan',3000,'E'],['PC','China',450,'F']]
tmp_frm = pd.DataFrame(data=data_lst, columns=cols)
tmp_frm
```

```python
tmp_frm.pivot(index='Machine',columns='Country',values='Price')
```

```python
tmp_frm.pivot(index='Machine', columns='Country',values=['Price','Brand'])
```

```python
tmp_frm.pivot(index='Machine',columns='Country')
```

```python
tmp_frm.pivot(index='Machine',columns='Country',values=['Price','Brand'])['Brand']
```

```python
tmp_frm.set_index(['Machine','Country'])[['Price','Brand']].unstack()
```

```python
data = {
    'Machine': ['A', 'A', 'B', 'B'],
    'Country': ['Korea', 'USA', 'Korea', 'USA'],
    'Price': [1000, 1500, 2000, 2500],
    'Brand': ['Samsung', 'Apple', 'LG', 'Microsoft']
}

df = pd.DataFrame(data)
df
# 'Machine' 'Country'  
df = df.set_index(['Machine', 'Country']).unstack()
df

```

```python
tmp_frm.set_index(['Machine','Country'])[['Price','Brand']].unstack()['Brand']
```

```python
tmp_frm.pivot(index=['Machine','Country'],columns='Brand',values='Price')
```

```python
tmp_frm.pivot(index='Country',columns=['Machine','Brand'],values='Price')
```
