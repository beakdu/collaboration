```python
import numpy as np
import pandas as pd
```

```python
df= pd.DataFrame({'월':[1,1,1,2,2,2],'이름':['a','b','c','a','b','c'],'야근':[1,2,3,0,2,5],'휴일':[8,4,1,1,3,4],'수당':[100,200,300,0,200,500]})
df
```

```python
df.apply(np.average,axis=0)
```

```python
df[['휴일','야근','수당']].apply(np.average,axis=0)
```

```python
df[['휴일','야근','수당']].apply(np.average, axis=1)
```

```python
def holiday_plus_overtime(df): return df['휴일']+df['야근']
```

```python
df['휴일과 야근']=df.apply(holiday_plus_overtime,axis=1)
df
```

```python


def is_overtime(overtime):
    if overtime <2:
        return "good"
    else:
        return "bad"
```

```python
df['초과야근여부']=df['야근'].map(lambda x:is_overtime(x))
df
```
