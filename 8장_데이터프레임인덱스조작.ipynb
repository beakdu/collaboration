```python
print('다중 인덱스 -열')
multi_frm =pd.DataFrame(np.round(np.random.randn(5,4),2),columns=[['Grp01','Grp01','Grp2','Grp2'],# 구릅인덱스,
                                                               ['col01','col02','col01','col02']]) # 컬럼인덱스
```

```python
multi_frm
```

```python
multi_frm.columns.names=['GrpIdx','ColIdx']
multi_frm # 그룹과 컬럼에 대한 인덱스 부여로 코드상 이점을 높일수 있음
```

```python
print('다중인덱스 -행')
multi_frm = pd.DataFrame(np.round(np.random.randn(6,4),2),columns=[['Grp01','Grp01','Grp02','Grp02'],
                                                                   ['col01','col02','col01','col02'],],
                         index=[
                             ['m','m','m','f','f','f'],
                             ['id_'+str(i+1) for i in range(6)]
                         ])
```

```python
multi_frm
```

```python
multi_frm.columns.names=['GrpIdx','ColIdx']
```

```python
multi_frm
```

```python
multi_frm.index.names=['gender','idx']
```

```python
multi_frm
```

```python
multi_frm.unstack()
```

```python
print('다중 인덱스의 경우 인덱싱은 어떻게?-')
print('튜플형식 접근하면 된다')
multi_frm[[('Grp01','col01'),('Grp02','col02')]]
```

```python
print('인덱스를 바꿀 수 있는데요-')
print('stack(): 열 ->행')
print('unstack():행 ->열')
```

```python
multi_frm.stack()
```

```python
# 행 인덱스 정보까지 같이 가져오기위해 loc사용
multi_frm.loc[[('m','id_1')],:] #m의 id_1번인 사람만 가져오기
```

```python
multi_frm.loc['m'] #m그룹에 대한 데이터정보 확인 가능
```

```python
print('다중 인덱스 정렬')
print('level 속성- 기준')
print('axis 축 지정 -')
multi_frm.sort_index(level=0)
multi_frm.sort_index(level=0, axis =1)
```
