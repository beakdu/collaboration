```python
import pandas as pd
```

```python
cols=['Machine','Country','Grade','Price','Count']
data_lst=[['TV','Korea','A',1000,3],['TV','Korea','B',800,8],['TV','Korea','B',800,2],
['TV','Japan','A',1300,5],['TV','Japan','A',1300,1],['PC','Korea','B',1500,6],['PC','Korea','A',2000,9],['PC','Japan','A',3000,3],['PC','Japan','B',2500,3]]
```

```python
pivot_table_frm=pd.DataFrame(data=data_lst,columns=cols)
```

```python
pivot_table_frm
```

```python
import numpy as np
```

```python
pivot_table_frm.pivot_table(index=['Machine','Country'],columns='Grade',values='Count',aggfunc=np.sum)
```

```python
pivot_table_frm.pivot_table(index=['Machine','Country'],columns='Grade',values=['Count','Price'],aggfunc=[np.sum,np.mean])
```

```python
pivot_table_frm.pivot_table(index=['Machine','Country'],columns='Grade',values='Count',aggfunc=np.sum,sort=True)
```

```python
pivot_table_frm.pivot_table(index=['Machine','Country'],columns='Grade',values=['Count','Price'],aggfunc={'Count':np.sum,'Price':np.mean})
```

```python
pivot_table_frm.pivot_table(index=['Machine','Country'],columns='Grade',values='Count',aggfunc=np.sum,fill_value='누락')
```

```python
pivot_table_frm.pivot_table(index=['Machine','Country'],columns='Grade',values='Count',aggfunc=np.sum, margins=True,margins_name='총계')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_datasets = sns.load_dataset('titanic')
titanic_datasets.head()
```

```python
print('성별과 객식등급에 따른 승객 수 집계한다면?-')
print('pivot-')
pclass_sex_frm=titanic_datasets.loc[:,['pclass','sex']]
pclass_sex_frm.head()
```

```python
pclass_sex_frm.groupby(['pclass','sex']).size()
```

```python
plcass_sex_pivot_frm=pclass_sex_frm.groupby(['pclass','sex']).size().reset_index(name='cnt')
plcass_sex_pivot_frm
```

```python
plcass_sex_pivot_frm.pivot(index='sex', columns='pclass', values='cnt')
```
