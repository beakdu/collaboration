```python
tmp_frm01=pd.DataFrame({'IDX1':['a','b','c','d'],'VAL':[1,2,3,4]})
tmp_frm01
```

```python
dict_a ={'id1':[1,2,3,4,5],'name':['a','b','c','d','e'],'price':[10,20,30,40,50]}
```

```python
dict_b={'id2':[1,2,3,4],'name':['a','b','z','z'],'price':[10,20,100,100]}
```

```python
join_frm01= pd.DataFrame(dict_a)
display(join_frm01)
print(join_frm01)
join_frm01
```

```python
join_frm02=pd.DataFrame(dict_b)
display(join_frm02)
```

```python
join_new_frm = join_frm01.join(join_frm02,lsuffix='_l',rsuffix='_r')
join_new_frm
```

```python
dict_a={'id':[1,2,3,4,5],'name':['a','b','c','d','e'],'price':[10,20,30,40,50]}
```

```python
dict_b={'id':[1,2,3,4],'name':['a','b','d','d'],'price':[10,20,100,100]}
join_frm03=pd.DataFrame(dict_a)
display(join_frm03)
join_frm04=pd.DataFrame(dict_b)
display(join_frm04)
```

```python
merge_new_frm=pd.DataFrame.merge(join_frm03,join_frm04, how='left',left_on=['id'],right_on=['id'],suffixes=['_','_r'])
merge_new_frm
```

```python
dict_a={'id1':[1,2,3,4,5],'name':['a','b','c','d','e'],'price':[10,20,30,40,50]}
dict_b={'id2':[1,2,3,4],'name':['a','b','z','z'],'price':[10,20,100,100]}
join_frm01=pd.DataFrame(dict_a)
join_frm02=pd.DataFrame(dict_b)
print(join_frm01)
print(join_frm02)
```

```python
join_new_frm=join_frm01.join(join_frm02, on=['id1'],lsuffix='_l',rsuffix='_r')
join_new_frm
```
