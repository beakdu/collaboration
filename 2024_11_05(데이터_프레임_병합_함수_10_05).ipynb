```python
import pandas as pd
```

```python
tmp_series01=pd.Series(['e0','e1','e2','e3'],name='e')
tmp_series02=pd.Series(['f0','f1','f2'],name='f',index=[3,4,5])
tmp_series03=pd.Series(['g0','g1','g2'],name='g')

print(tmp_series01)
print(tmp_series02)
print(tmp_series03)
```

```python
result01=pd.concat([tmp_series01,tmp_series02],axis=0)
print(result01)
```

```python
result02=pd.concat([tmp_series01,tmp_series02],axis=1)
print(result02)
```

```python
tmp_frm01=pd.DataFrame({'IDX1':['a','b','c','d'],'VAL':[1,2,3,4]})
tmp_frm01
```

```python
tmp_frm02=pd.DataFrame({'IDX2':['a','b','c'],'VAL':[5,6,7]})
tmp_frm02
```

```python
tmp_frm01.merge(tmp_frm02, left_on='IDX1',right_on='IDX2')
```

```python
tmp_frm02.merge(tmp_frm01, left_on='IDX2',right_on='IDX1')
```

```python
result02=pd.concat([tmp_frm01,tmp_frm02],axis=1)
print(result02)
```

```python
tmp_frm01.merge(tmp_frm02, left_on='IDX1',right_on='IDX2',suffixes=('_left','_right'))
```

```python
tmp_frm03=pd.DataFrame({'VAL1':[1,2,3]},index=['row1','row2','row3'])
tmp_frm03
```

```python
tmp_frm04=pd.DataFrame({'VAL2':[4,5,6]},index=['row2','row3','row4'])
tmp_frm04
```

```python
tmp_frm03.merge(tmp_frm04,left_index=True, right_index=True)
```

```python
tmp_frm05=pd.DataFrame({'VAL1':[1,2,3]},index=['row1','row2','row3'])
tmp_frm05
```

```python
tmp_frm06=pd.DataFrame({'IDX':['row2','row3','row4'],'VAL2':[4,5,6]})
tmp_frm06
```

```python
tmp_frm05.merge(tmp_frm06, left_index=True, right_on='IDX')
```

```python
tmp_frm07=pd.DataFrame({'IDX':['a','b','c','a'],'VAL':[1,2,3,4]})
tmp_frm07
```

```python
tmp_frm08=pd.DataFrame({'IDX':['row2','row3','row4'],'VAL2':[4,5,6]})
tmp_frm08
```

```python
tmp_frm07.merge(tmp_frm08, how='left', on='IDX')
```
