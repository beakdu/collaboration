```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic_dataset = sns.load_dataset('titanic')
titanic_dataset.head()
```

```python
print('선실 등급에 따른 그룹 -')
titanic_subset_frm = titanic_dataset.loc[:,['age','sex','class','fare','survived']]
titanic_subset_frm.head()
```

```python
class_grp= titanic_subset_frm.groupby(['class'])
class_grp
```

```python
for key,group in class_grp:
    print('key-',key)

    print('len',len(group))
    print('group-',group)
```
